```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Define parameters
original_sample_size = 179
simulated_sample_sizes = [30, 50, 100, 125]  # Different sample sizes for simulation
n_simulations = 1000
true_correlation = 0.247025677  # Assumed true correlation in the simulated data

# Function to simulate data and calculate correlation and p-value
def simulate_and_calculate_correlation(sample_size, true_corr, n_sim):
    correlations = []
    p_values = []
    for _ in range(n_sim):
        # Simulate data
        x = np.random.randn(sample_size)
        y = true_corr * x + np.sqrt(1 - true_corr**2) * np.random.randn(sample_size)
        # Calculate correlation and p-value
        corr, p_value = pearsonr(x, y)
        correlations.append(corr)
        p_values.append(p_value)
    return correlations, p_values

# Perform simulations for different sample sizes
results = {}
for size in simulated_sample_sizes:
    correlations, p_values = simulate_and_calculate_correlation(size, true_correlation, n_simulations)
    results[size] = {'correlations': correlations, 'p_values': p_values}

# Analyze and plot results
for size, result in results.items():
    correlations = result['correlations']
    p_values = result['p_values']
    
    mean_corr = np.mean(correlations)
    ci_corr = np.percentile(correlations, [2.5, 97.5])
    mean_p_value = np.mean(p_values)
    ci_p_value = np.percentile(p_values, [2.5, 97.5])
    
    print(f"Sample Size {size}:")
    print(f"  Mean Correlation = {mean_corr}")
    print(f"  95% CI for Correlation = {ci_corr}")
    print(f"  Mean p-value = {mean_p_value}")
    print(f"  95% CI for p-value = {ci_p_value}")

    # Plot histogram of correlations
    sns.histplot(correlations, kde=True)
    plt.title(f'Simulation Results for Sample Size {size} - Correlations')
    plt.xlabel('Correlation')
    plt.show()

    # Plot histogram of p-values
    sns.histplot(p_values, kde=True, bins=20)
    plt.title(f'Simulation Results for Sample Size {size} - p-values')
    plt.xlabel('p-value')
    plt.show()
```

```python
#power analysis
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from statsmodels.stats.power import NormalIndPower

# Given data
r_value = 0.24
n = 179  # sample size
alpha = 0.05  # significance level

# Function for power analysis
def perform_power_analysis(r, n, alpha=0.05):
    # Convert correlation to effect size (Cohen's d equivalent for correlation)
    effect_size = np.sqrt(r**2 / (1 - r**2))
    
    # Perform power analysis
    power_analysis = NormalIndPower()
    power = power_analysis.solve_power(effect_size=effect_size, nobs1=n, alpha=alpha, alternative='two-sided')
    
    return power

# Function to plot confidence interval
def plot_confidence_interval(r, n, alpha=0.05):
    # Calculate standard error of the correlation coefficient
    se = 1 / np.sqrt(n - 3)
    
    # Fisher's z-transform to calculate confidence interval
    z = np.arctanh(r)
    z_critical = 1.96  # For 95% CI
    
    # Confidence interval in z-space
    z_lower = z - z_critical * se
    z_upper = z + z_critical * se
    
    # Transform back to r-space
    r_lower = np.tanh(z_lower)
    r_upper = np.tanh(z_upper)
    
    # Plot the correlation with its confidence interval
    plt.figure(figsize=(8, 6))
    plt.errorbar(r, 0, xerr=[[r - r_lower], [r_upper - r]], fmt='o', capsize=5)
    plt.title(f'Correlation: {r}, 95% CI: [{r_lower:.3f}, {r_upper:.3f}]')
    plt.xlabel('Correlation coefficient')
    plt.show()
    
    return r_lower, r_upper

# Perform power analysis and plot confidence interval
power = perform_power_analysis(r_value, n, alpha)
confidence_interval = plot_confidence_interval(r_value, n)

print(f"Power: {power}")
print(f"95% Confidence Interval: {confidence_interval}")
```
